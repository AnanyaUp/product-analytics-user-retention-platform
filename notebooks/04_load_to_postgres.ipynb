```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

USERNAME = "postgres"
PASSWORD = quote_plus("Ananya@05")
HOST = "localhost"
PORT = "5432"
DATABASE = "product_analytics"

engine = create_engine(
    f"postgresql+psycopg2://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
)

print("Connection Object Created Successfully")
```

```python
import pandas as pd

sample_df = pd.read_csv(
    "../data/processed/events_sample.csv"
)

print(sample_df.shape)
```

```python
# Convert DateTime
sample_df["event_time"] = pd.to_datetime(
    sample_df["event_time"]
)
```

```python
# Upload to PostgreSQL
sample_df.to_sql(
    "events",
    engine,
    if_exists="append",
    index=False,
    chunksize=10000,
    method="multi"
)

print("Upload Successful")
```
