```python
import pandas as pd

file_path = "../data/raw/2019-Nov/2019-Nov.csv"

sample_df = pd.read_csv(
    file_path,
    skiprows=lambda x: x > 0 and x % 100 != 0
)

sample_df.shape
```

```python
sample_df["event_time"] = pd.to_datetime(sample_df["event_time"])

print(sample_df["event_time"].min())
print(sample_df["event_time"].max())
```

```python
sample_df["event_type"].value_counts()
```

```python
# Create processed folder if it doesn't exist
import os

os.makedirs("../data/processed", exist_ok=True)

# Save sample dataset
sample_df.to_csv(
    "../data/processed/events_sample.csv",
    index=False
)

print("events_sample.csv saved successfully!")
```
